```python
import sys
import pathlib

ROOT = pathlib.Path(".").absolute().parent.parent
sys.path.append(str(ROOT))

import os
import numpy as np
from models.raw_audio import ResTSSDNetWrapper
from datasets.SPCUP22DataModule import SPCUP22DataModule

from utils.metrics import (
    plot_classification_report,
    pytorch_lightning_make_predictions as make_predictions,
    print_scores,
    plot_tsne_features
)
```

### Change this according to your system

```python
os.environ["CUDA_VISIBLE_DEVICES"] = "6"
```

### Without augmentations

```python
checkpoint = ResTSSDNetWrapper.load_from_checkpoint(
    ROOT.joinpath("checkpoints", "tssdnet", "res_tssdnet_with_unseen", "last.ckpt"),
    return_features=True
)

data_module = SPCUP22DataModule(
    64,
    dataset_root=pathlib.Path(ROOT).joinpath("data", "raw_audio", "spcup22"),
    should_include_unseen_in_training_data=True,
    num_workers=8,
)
```

```python
actual_labels, predictions, probabilities, _, final_layer_features = make_predictions(
    checkpoint, data_module, return_final_layer_features=True
);
```

```python
print_scores(actual_labels, predictions)
```

```python
fig_save_path = str(ROOT.joinpath("results", "tssdnet", "res", "with_unseen"))
os.makedirs(fig_save_path, exist_ok=True)

plot_classification_report(
    actual_labels,
    predictions,
    probabilities,
    title_suffix="Res-TSSDNet (With unseen samples)",
    save_path=fig_save_path
)
```

```python
final_layer_features = np.array(final_layer_features, dtype=np.float32)

fig, ax = plot_tsne_features(
    final_layer_features,
    actual_labels,
    "t-SNE for Res-TSSDNet (no augmentations)",
    use_grey=False
)

fig.savefig(
    ROOT.joinpath("results", "res_tssdnet_no_aug.eps"),
    format="eps",
    bbox_inches="tight",
)
```

### With augmentations

```python
checkpoint = ResTSSDNetWrapper.load_from_checkpoint(
    ROOT.joinpath("checkpoints", "tssdnet", "res_tssdnet_with_unseen_aug", "last.ckpt"),
    return_features=True
)

data_module = SPCUP22DataModule(
    64,
    dataset_root=pathlib.Path(ROOT).joinpath("data", "raw_audio", "spcup22"),
    should_include_unseen_in_training_data=True,
    should_include_augmented_data=True,
    num_workers=8,
)
```

```python
actual_labels, predictions, probabilities, _, final_layer_features = make_predictions(
    checkpoint, data_module, return_final_layer_features=True
);
```

```python
print_scores(actual_labels, predictions)
```

```python
fig_save_path = str(ROOT.joinpath("results", "tssdnet", "res", "with_unseen_aug"))
os.makedirs(fig_save_path, exist_ok=True)

plot_classification_report(
    actual_labels,
    predictions,
    probabilities,
    title_suffix="Res-TSSDNet (With unseen, augmented samples)",
    save_path=fig_save_path
)
```

```python
final_layer_features = np.array(final_layer_features, dtype=np.float32)

fig, ax = plot_tsne_features(
    final_layer_features,
    actual_labels,
    "t-SNE for Res-TSSDNet (with augmentations)",
    use_grey=False
)

fig.savefig(
    ROOT.joinpath("results", "res_tssdnet_aug.eps"),
    format="eps",
    bbox_inches="tight",
)
```
